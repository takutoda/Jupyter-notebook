- DaruでCSVファイルの読み込み

```ruby
require 'daru'

df = Daru::DataFrame.from_csv("https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/anime_master.csv", opts = {encoding: "ISO8859-1:utf-8"})
```

```ruby
# 基本統計
df.describe
```

- Nyaplotのチャート例

```ruby
pivoted = df.pivot_table(index: ['type'], vectors: [], agg: :sum, values: :members )
pivoted["type"] = pivoted.index
pivoted.plot(type: :bar, x: :type, y: :members) do |p, d|
  p.x_label 'Type'
  p.y_label 'Members'
  p.margin({top:30, left: 150, bottom:80, right:30})
end
```
